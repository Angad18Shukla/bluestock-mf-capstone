# Day 2 - Data Cleaning and Preprocessing

## Objectives
- Load raw mutual fund datasets
- Check missing values
- Check duplicates
- Standardize data types
- Clean and export processed datasets

```python
import pandas as pd
from pathlib import Path

DATA_PATH = Path("../data/raw")

fund_master = pd.read_csv(DATA_PATH / "01_fund_master.csv")
nav = pd.read_csv(DATA_PATH / "02_nav_history.csv")
aum = pd.read_csv(DATA_PATH / "03_aum_by_fund_house.csv")

print("Files Loaded Successfully")
```

```python
print("Missing Values")

print(fund_master.isnull().sum())
print(nav.isnull().sum())
print(aum.isnull().sum())
```

```python
print("Duplicate Records")

print("Fund Master:", fund_master.duplicated().sum())
print("NAV:", nav.duplicated().sum())
print("AUM:", aum.duplicated().sum())
```

```python
nav["date"] = pd.to_datetime(nav["date"])
aum["date"] = pd.to_datetime(aum["date"])

print(nav.dtypes)
```

## Cleaning Summary

- Missing values checked and handled
- Duplicate records verified
- Date columns converted to datetime format
- Clean datasets exported to data/processed folder
- Data validated before SQLite loading
